#DATA CLEANING AND TRANSSFORMATION

```python
##drop missing value as seen that orderID and customer ID and customer name have 20 missing value each
yanki_df.dropna(subset=['Order_ID', 'Customer_ID'], inplace=True)
```

```python
yanki_df.info()
```

```python
#convert order date from string to datetime object
yanki_df['Order_Date'] = pd.to_datetime(yanki_df['Order_Date'], format='%d/%m/%Y %H:%M')
```

```python
yanki_df.columns
```

```python
#customer table

customer_df=yanki_df[['Customer_ID','Customer_Name','Email','Phone_Number']].copy().drop_duplicates().reset_index(drop=True)


#create a customer table and because it will be a subset, to eradicate duplictae we drop duplicate and and index attached to it 
#and return it to be same table instead of creating a new table
```

```python
customer_df
```

```python
#product table
product_df=yanki_df[['Product_Name','Brand','Category','Price']].copy().drop_duplicates().reset_index(drop=True)
```

```python
product_df
```

```python
#shipping addreess table
shipping_address_df=yanki_df[['Customer_ID','Shipping_Address', 'City', 'State', 'Country',
       'Postal_Code']].copy().drop_duplicates().reset_index(drop=True)

shipping_address_df.index.name='shipping_ID'
shipping_address_df=shipping_address_df.reset_index() 

#add anew column to dataframe to name each index(index is like Id to each column) #and name it shippingID

shipping_address_df
```

```python
#order date table
orders_df = yanki_df[['Order_ID','Customer_ID','Product_ID','Quantity', 'Total_Price',
       'Order_Date']].copy().drop_duplicates().reset_index(drop=True)

orders_df.head()
```

```python
##payment_method table
payment_method_df=yanki_df[['Order_ID','Payment_Method',
       'Transaction_Status']].copy().drop_duplicates().reset_index(drop=True)
payment_method_df
```
